```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('./anomaly_detection_dataset_2022/trainset.csv')
df_anomaly = pd.read_csv('./anomaly_detection_dataset_2022/anomaly_testset1.csv')

df_train['created_at'] = pd.to_datetime(df_train['created_at'])
df_anomaly['created_at'] = pd.to_datetime(df_anomaly['created_at'])
df_train.size, df_anomaly.size
```

```python
for col in df_train.columns.tolist():
    if (col == 'created_at'):
        continue
    plt.figure(figsize=(20, 3))
    plt.title(col)
    plt.plot(df_train['created_at'], df_train[col])
    plt.show()
```

```python
for col in df_anomaly.columns.tolist():
    if (col == 'created_at'):
        continue
    plt.figure(figsize=(20, 3))
    plt.title(col)
    plt.plot(df_anomaly['created_at'], df_anomaly[col])
    plt.show()
```
