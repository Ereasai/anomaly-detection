```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('./anomaly_detection_dataset_2022/trainset.csv')
df_anomaly = pd.read_csv('./anomaly_detection_dataset_2022/anomaly_testset1.csv')

df_train['created_at'] = pd.to_datetime(df_train['created_at'])
df_anomaly['created_at'] = pd.to_datetime(df_anomaly['created_at'])
```

```python
df_train_features = df_train.drop(columns=['created_at'])
df_anomaly_truncated = df_anomaly.drop(columns=['created_at', 'scene', 'label'])
```

```python
mean = df_train_features.mean().to_numpy()
std = df_train_features.mean().to_numpy()
```

```python
mean, std
```

```python
thresh_min = mean - 3 * std
thresh_max = mean + 3 * std
```

```python
anomalies = (df_train_features < thresh_min) | (df_train_features > thresh_max)
```

```python
row_with_anomalies = anomalies.any(axis=1)
df_anomaly_truncated = df_anomaly_truncated[row_with_anomalies]
df_anomaly_truncated
```
